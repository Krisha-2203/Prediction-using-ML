```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,1:].values
```

```python
x
```

```python
y
```

```python
#Training the model
#Import required libraries
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
```

```python
#Using 80% of the data for training and 20% for testing
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Visualizing the trianed model
line = regression.coef_*x+regression.intercept_

plt.scatter(x,y)
plt.plot(x, line, color='blue')
plt.title("Student Score")
plt.xlabel("Study Hours")
plt.ylabel("Percentage Scores")
plt.grid()
plt.show()
```

```python
#Prediction of the data
y_pred = regression.predict(x_test)
```

```python
datapredict = {'Actual': [y_test], 'Predicted':[y_pred]}
dframe = pd.DataFrame(datapredict)
dframe
```

```python
#Predict score of the student if he studies for 9.25 hours
hours = 9.25
own_pred = regression.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Scores = {}".format(own_pred[0]))
```

```python
#Model Evaluation
from sklearn import metrics
from sklearn.metrics import r2_score
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))
```
